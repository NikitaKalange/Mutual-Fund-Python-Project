# Mutual Fund Plan with Python

### Importing libraries & Loading file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("nifty50_closing_prices - nifty50_closing_prices.csv")
df.head()
```

### Convert Date Column

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.head()
```

### Check Missing Values

```python
df.isnull().sum()
```

### Calculate Daily Returns

```python
returns = df.pct_change().dropna()
```

### Annualized Return

```python
annual_returns = returns.mean() * 252
```

### Annualized Volatility

```python
annual_volatility = returns.std() * np.sqrt(252)         # Standard deviation measures risk, Higher std = more fluctuations
```

### Risk-Return DataFrame

```python
risk_return_df = pd.DataFrame({
    'Returns': annual_returns,
    'Volatility': annual_volatility
})
```

## Exploratory Data Analysis (EDA)

### Heatmap of Correlation

### Risk vs Return Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(annual_volatility, annual_returns)
plt.xlabel("Volatility")
plt.ylabel("Returns")
plt.title("Risk vs Return")
plt.show()                             # Top-left is best (high return, low risk)
```

```python
We want to combine multiple stocks into one portfolio and calculate:
1. Portfolio Return
2. Portfolio Risk (Volatility)
```

### Decide Portfolio Weights

```python
selected_stocks = returns.columns[:3]                       # automatically pick first 3 stocks for simplicity
selected_returns = returns[selected_stocks]
weights = np.array([0.4, 0.3, 0.3])                         # Define portfolio weights manually
```

### Calculate Portfolio Return

```python
# Simple Formula - PortfolioReturn = (w1×r1)+(w2×r2)+(w3×r3)

annual_returns_selected = selected_returns.mean() * 252               # Calculate average annual return of selected stocks
portfolio_return = np.sum(weights * annual_returns_selected)          # Portfolio expected annual return
print("Portfolio Annual Return:", portfolio_return)
```

### Calculate Portfolio Risk (Volatility)

```python
covariance_matrix = selected_returns.cov() * 252

weights_matrix = np.reshape(weights, (len(weights), 1))             # Convert weights into column format

portfolio_variance = np.dot(weights_matrix.T, np.dot(covariance_matrix, weights_matrix))
portfolio_volatility = np.sqrt(portfolio_variance)

print("Portfolio Risk (Volatility):", portfolio_volatility[0][0])
```

## Monte Carlo Simulation

### Create Empty Lists

```python
portfolio_returns = []
portfolio_risks = []
portfolio_sharpes = []
portfolio_weights = []
```

### Run 20 Portfolio Combinations

```python
risk_free_rate = 0.06         # assume 6% risk-free rate

for i in range(20):
    weights = np.random.random(len(selected_stocks))        # Generate random weights

    weights = weights / np.sum(weights)                     # Make sure total weight = 1

    portfolio_return = np.sum(weights * annual_returns_selected)

    portfolio_variance = np.dot(weights.T, np.dot(covariance_matrix, weights))
    portfolio_risk = np.sqrt(portfolio_variance)

    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_risk

    # Store values
    portfolio_returns.append(portfolio_return)
    portfolio_risks.append(portfolio_risk)
    portfolio_sharpes.append(sharpe_ratio)
    portfolio_weights.append(weights)
```

### Convert Lists to DataFrame

```python
results_df = pd.DataFrame({
    "Return": portfolio_returns,
    "Risk": portfolio_risks,
    "Sharpe Ratio": portfolio_sharpes
})

results_df
```

### Find Best Portfolio

```python
best_index = results_df["Sharpe Ratio"].idxmax()

best_portfolio = results_df.loc[best_index]

print("Best Portfolio:")
print(best_portfolio)

print("\nBest Weights Allocation:")
print(portfolio_weights[best_index])
```

### Simple Risk-Return Plot

#### Highlight Best Portfolio on Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(portfolio_risks, portfolio_returns)

plt.scatter(portfolio_risks[best_index],
            portfolio_returns[best_index],
            marker='*', s=300)

plt.xlabel("Risk")
plt.ylabel("Return")
plt.title("20 Portfolio Risk-Return Comparison")

plt.show()
```
